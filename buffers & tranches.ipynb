```python
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
# define a random variable representing cash inflow
sample = norm.rvs(size=10*12,loc=1000,scale=1000)
sample = [max(0, elem) for elem in sample]
sample = [min(2000, elem) for elem in sample]
```

```python
# plot cach inflow
plt.plot(sample)
plt.show()
```

```python
# consider using zip function
def s_buffer(buffer_input, buffer_output, initial, buffer_size):
    toReturn = []
    myBuffer = []
    for sample in buffer_input:
        initial = min(initial, buffer_size)
        myBuffer.append(initial)
        initial += sample
        toAppend = min(initial, buffer_output)
        initial -= toAppend
        toReturn.append(toAppend)
    return [toReturn, myBuffer]
```

```python
# constant cash outflow @ $1000
exzero = s_buffer(sample, 1000, 0, 0)
plt.plot(exzero[0])
plt.plot(exzero[1])
plt.show()
```

```python
# constant cash outflow @ $1000
exone = s_buffer(sample, 1000, 1000, 5000)
plt.plot(exone[0])
plt.plot(exone[1])
plt.show()
```

```python
# constant cash outflow @ $1000
extwo = s_buffer(sample, 1000, 2500, 5000)
plt.plot(extwo[0])
plt.plot(extwo[1])
plt.show()
```

```python
# constant cash outflow @ $1000
exthree = s_buffer(sample, 1000, 1000, 10000)
plt.plot(exthree[0])
plt.plot(exthree[1])
plt.show()
```
